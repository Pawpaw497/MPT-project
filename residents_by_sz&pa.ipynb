```python
#population_subzones
#author: pawpaw
#information included in the source date: planning_area,subzone,age_group,sex,type_of_dwelling,resident_count,year
# year=2019, resident number
import os 
import csv 
import pandas as pd 
from pandas.core import groupby
```

```python
path = '/Users/paw_paw/Library/CloudStorage/GoogleDrive-keyuqing597@gmail.com/我的云端硬盘/PC5287_M.SC Coursework Thesis for Physics And Technology/data set'
residents = path+'/singapore-residents-by-subzone-and-type-of-dwelling-2011-2019/planning-area-subzone-age-group-sex-and-type-of-dwelling-june-2011-2019.csv'
df_res = pd.read_csv(residents)
df_res
```

```python
df_res = df_res[df_res['year'].isin([2019])]
# select the data of 2019
df_res
```

```python
res_sub = df_res.groupby(["subzone"]).sum(numeric_only=True)["resident_count"].reset_index()
res_sub
```

```python
residents_by_subzone = path+'/singapore_residents_by_subzone2019.csv'
res_sub.to_csv(residents_by_subzone) 
```

```python
res_pa = df_res.groupby(["planning_area"]).sum(numeric_only=True)["resident_count"].reset_index()
res_pa
```

```python
residents_by_pa = path+'/singapore_residents_by_pa2019.csv'
res_pa.to_csv(residents_by_pa) 
```
