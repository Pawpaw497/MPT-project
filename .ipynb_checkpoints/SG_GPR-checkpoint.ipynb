```python
def set_precision(geo_list):
    for element in geo_list:
        element[0] = ("%.14f"%element[0])
        element[1] = ("%.14f"%element[1])
    return geo_list
```

```python
data1.Geo=data1.Geo.apply(set_precision)
```

```python
def get_polygon(point_list):
    temp_df = pd.DataFrame(point_list)
    temp_gdf = gpd.GeoDataFrame(temp_df, geometry = gpd.points_from_xy(temp_df[0],temp_df[1]))
    poly = geometry.Polygon([[p.x, p.y] for p in temp_gdf.geometry])
    #https://stackoverflow.com/questions/30457089/how-to-create-a-shapely-polygon-from-a-list-of-shapely-points
    return poly 
```

```python
data1['ply'] = data1.Geo.apply(get_polygon)
```

```python
for PA in Prediction.columns[0:]:
    def prediction(pa1,pa2=PA):
        temp = R_ij.set_index(['PA_name'])[pa1].to_dict()
        r = temp.get(pa2)
        s_ij = 0
        centroid = GPR['centroid'].to_dict()
        POP_DENS = GPR['POP_DENS'].to_dict()
        for zone in range(0,33117): 
            if centroid.get(zone).distance(pa_dict.get(PA))<= r:
                s_ij = s_ij + POP_DENS.get(zone)
        s_ij = s_ij - m_dict.get(pa1)
        
        def get_Ti_bus(pa):
            if T_i_bus_dict.get(pa1) is None:
                return float(0)
            else:
                return T_i_bus_dict.get(pa1)
        def get_Ti_MRT(pa):
            if T_i_MRT_dict.get(pa1) is None:
                return float(0)
            else:
                return T_i_MRT_dict.get(pa1)
        T_i = get_Ti_bus(pa1) + get_Ti_MRT(pa1)
        if ((m_dict.get(pa1)+s_ij)*(m_dict.get(pa1)+s_ij+m_dict.get(pa2)))== 0:
            pre = None
        else:
            pre = T_i*m_dict.get(pa1)*m_dict.get(pa2)/((m_dict.get(pa1)+s_ij)*(m_dict.get(pa1)+s_ij+m_dict.get(pa2)))
        return pre 
    Prediction[PA]=pa['PLN_AREA_N'].apply(prediction)    

Prediction
```

```python
new = np.empty([55,55])

def bus_flux(x,y):
    pa1 = Prediction.columns[x]
    pa2 = Prediction.columns[y]
    
    if T_ij_bus[(T_ij_bus['ORIGIN_loc']==pa1)&(T_ij_bus['DESTINATION_loc']==pa2)].size==0:
        return int(0)
    else:
        flux = T_ij_bus[(T_ij_bus['ORIGIN_loc']==pa1)&(T_ij_bus['DESTINATION_loc']==pa2)]['TOTAL_TRIPS'].values[0]
        return flux

for i in range(0,55):
    for j in range(0,55):
        new[i][j]=bus_flux(i,j)
        new[i][i]=0

bus_flux_array = new

new = np.empty([55,55])
def MRT_flux(x,y):
    pa1 = Prediction.columns[x]
    pa2 = Prediction.columns[y]
    
    if T_ij_MRT[(T_ij_MRT['ORIGIN_loc']==pa1)&(T_ij_MRT['DESTINATION_loc']==pa2)].size==0:
        return int(0)
    else:
        flux = T_ij_MRT[(T_ij_MRT['ORIGIN_loc']==pa1)&(T_ij_MRT['DESTINATION_loc']==pa2)]['TOTAL_TRIPS'].values[0]
        return flux
for i in range(0,55):
    for j in range(0,55):
        new[i][j]=MRT_flux(i,j)
MRT_flux_array = new

flux_array=(bus_flux_array+MRT_flux_array)

flux = pd.DataFrame(flux_array)
```
