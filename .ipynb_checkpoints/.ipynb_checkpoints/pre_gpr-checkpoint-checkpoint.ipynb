$$
\langle T_{ij} \rangle = T_i\cdot \frac{m_i\cdot n_j}{(m_i+s_{ij})(m_i+n_j+s_{ij})}
$$
in which:
$$
T_i=\sum_{j\neq i}T_{ij}
$$
and $\langle T_{ij} \rangle$ means tha mean value of $T_{ij}$ within a period of time.

```python
new = np.empty([55,55])

def bus_flux(x,y):
    pa1 = Prediction.columns[x]
    pa2 = Prediction.columns[y]
    
    if T_ij_bus[(T_ij_bus['ORIGIN_loc']==pa1)&(T_ij_bus['DESTINATION_loc']==pa2)].size==0:
        return int(0)
    else:
        flux = T_ij_bus[(T_ij_bus['ORIGIN_loc']==pa1)&(T_ij_bus['DESTINATION_loc']==pa2)]['TOTAL_TRIPS'].values[0]
        return flux

for i in range(0,55):
    for j in range(0,55):
        new[i][j]=bus_flux(i,j)
        new[i][i]=0

bus_flux_array = new

new = np.empty([55,55])
def MRT_flux(x,y):
    pa1 = Prediction.columns[x]
    pa2 = Prediction.columns[y]
    
    if T_ij_MRT[(T_ij_MRT['ORIGIN_loc']==pa1)&(T_ij_MRT['DESTINATION_loc']==pa2)].size==0:
        return int(0)
    else:
        flux = T_ij_MRT[(T_ij_MRT['ORIGIN_loc']==pa1)&(T_ij_MRT['DESTINATION_loc']==pa2)]['TOTAL_TRIPS'].values[0]
        return flux
for i in range(0,55):
    for j in range(0,55):
        new[i][j]=MRT_flux(i,j)
MRT_flux_array = new

flux_array=(bus_flux_array+MRT_flux_array)

flux = pd.DataFrame(flux_array)
```
