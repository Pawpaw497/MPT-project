```python
new = np.empty([55,55])

def bus_flux(x,y):
    pa1 = Prediction['PA_name'][x]
    pa2 = Prediction['PA_name'][y]
    
    if T_ij_bus[(T_ij_bus['ORIGIN_loc']==pa1)&(T_ij_bus['DESTINATION_loc']==pa2)].size==0:
        return int(0)
    else:
        flux = T_ij_bus[(T_ij_bus['ORIGIN_loc']==pa1)&(T_ij_bus['DESTINATION_loc']==pa2)]['TOTAL_TRIPS'].values[0]
        return flux

for i in range(0,55):
    for j in range(0,55):
        new[i][j]=bus_flux(i,j)
        new[i][i]=0

bus_flux_array = new

new = np.empty([55,55])
def MRT_flux(x,y):
    pa1 = Prediction['PA_name'][x]
    pa2 = Prediction['PA_name'][y]
    
    if T_ij_MRT[(T_ij_MRT['ORIGIN_loc']==pa1)&(T_ij_MRT['DESTINATION_loc']==pa2)].size==0:
        return int(0)
    else:
        flux = T_ij_MRT[(T_ij_MRT['ORIGIN_loc']==pa1)&(T_ij_MRT['DESTINATION_loc']==pa2)]['TOTAL_TRIPS'].values[0]
        return flux
for i in range(0,55):
    for j in range(0,55):
        new[i][j]=MRT_flux(i,j)
MRT_flux_array = new

flux_array=(bus_flux_array+MRT_flux_array)

flux = pd.DataFrame(flux_array)
```

```python
new_stat = pd.DataFrame()
new_stat['PLA'] = Prediction.columns

for i in range(0,55):
    cleanedList = [x for x in Prediction.to_numpy()[i] if np.isnan(x) == False]
    new_stat.loc[i,'mean'] = statistics.mean(cleanedList)
    new_stat.loc[i,'std_dvt'] = np.std(cleanedList)

new_stat
```
