```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
```

```python
path  = 'G:\我的云端硬盘\PC5287_M.SC Coursework Thesis for Physics And Technology\data set'
```

```python
R_ij = pd.read_csv(path +'\distance_pa_to_pa.csv').drop(columns='Unnamed: 0')

R_circle = pd.read_csv(path +'\dist_pa_to_sz.csv').drop(columns='Unnamed: 0')
```

```python
m = pd.read_csv(path + '\singapore_residents_by_pa2019.csv').drop(columns='Unnamed: 0')

s = pd.read_csv(path + '\singapore_residents_by_subzone2019.csv').drop(columns='Unnamed: 0')
```

```python
T_i_bus = pd.read_csv(path + '\Total_bus_flux2019.csv').drop(columns='Unnamed: 0')

T_ij_bus= pd.read_csv(path+'\Total_bus_flux_ij2019.csv').drop(columns='Unnamed: 0')
```

```python
T_i_MRT = pd.read_csv(path + '\Total_MRT_flux2019.csv').drop(columns='Unnamed: 0')

T_ij_MRT = pd.read_csv(path + '\Total_MRT_flux_ij2019.csv').drop(columns='Unnamed: 0')
```

```python
T_i_bus_dict = T_i_bus.set_index(['ORIGIN_loc'])['TOTAL_TRIPS'].to_dict()

T_i_MRT_dict= T_i_MRT.set_index(['ORIGIN_loc'])['TOTAL_TRIPS'].to_dict()
```

```python
s_dict = s.set_index(['subzone'])['resident_count'].to_dict()

m_dict = m.set_index(['planning_area'])['resident_count'].to_dict()
```

```python
R_circle_dict = R_circle.set_index(['SZ_name'])['BISHAN'].to_dict()
```

```python
Prediction = R_ij
```

```python
for pa in Prediction.columns[1:]:
    def prediction(pa1,pa2=pa):
        temp = R_ij.set_index(['PA_name'])[pa1].to_dict()
        r = temp.get(pa2)
        R_circle_dict = R_circle.set_index(['SZ_name'])[pa1].to_dict()
        s_ij = 0
        for sz in R_circle_dict.keys(): 
            if R_circle_dict.get(sz)<= r:
                s_ij = s_ij + s_dict.get(sz)
        s_ij = s_ij - m_dict.get(pa1)
        
        def get_Ti_bus(pa):
            if T_i_bus_dict.get(pa1) is None:
                return float(0)
            else:
                return T_i_bus_dict.get(pa1)
        def get_Ti_MRT(pa):
            if T_i_MRT_dict.get(pa1) is None:
                return float(0)
            else:
                return T_i_MRT_dict.get(pa1)
        T_i = get_Ti_bus(pa1) + get_Ti_MRT(pa1)
        if ((m_dict.get(pa1)+s_ij)*(m_dict.get(pa1)+s_ij+m_dict.get(pa2)))==0:
            pre = None
        else:
            pre = T_i*m_dict.get(pa1)*m_dict.get(pa2)/((m_dict.get(pa1)+s_ij)*(m_dict.get(pa1)+s_ij+m_dict.get(pa2)))
        return pre 
    Prediction[pa]=Prediction['PA_name'].apply(prediction)
```
